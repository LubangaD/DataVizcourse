# Correlation Visuals


- Correlation is a statistical measure of the relationship between two variables that demonstrate a linear relationship between each other. 

- It represents how closely two variables are connected. 

1. Positive correlation - variables tends to move in the same direction. eg. when one variable increases, the other variable increases.

2. Negative correlation - variables tend to move in opposite directions eg. when one variable increases, the other decreases

3. No correlation - the variables do not have a relationship with each other.







### Scatter plot

Scatter plots are great for plotting two variables to visualize how they might correlate and what relationship there mighat be between the variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head()
```

```python


# Create a scatter plot using Seaborn
sns.scatterplot(data=tips_df, x="total_bill", y="tip")

# Set plot title and axis labels
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

```python

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

## Joint Plot


Joint plots help us understand the relationship between two variables and the distribution of each variable.

It consists of 3 seperate plots whereby the middle figure shows the relationship between x and y, while the two other plots are histograms 

```python
sns.jointplot(data=tips_df, x="total_bill", y="tip")
```

```python

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a joint plot
sns.jointplot(data=iris, x="sepal_length", y="sepal_width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

# HeatMap

Heatmap is a useful chart that we can use to show the relationship between two variables. Heatmap displays a general view of numerical data; it does not extract specific data points. It is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
tips_df
```

```python
# get the correlation using the pandas corr() method
corr_df = tips_df.corr(numeric_only=True)
```

```python
corr_df
```

```python
sns.heatmap(corr_df,annot=True,cmap='flare')
```

```python
# Heatmap to see correlation in the mpg dataset
mpg = sns.load_dataset('mpg')
mpg
```

```python
corr_df2 = mpg.corr(numeric_only=True)
```

```python
corr_df2
```

```python
sns.heatmap(corr_df2, annot=True, cmap="Oranges")
```
