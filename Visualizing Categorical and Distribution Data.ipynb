# Visualizations with Distributions

Visualizing distributions allows you to understand critical statistical properties of the data you have.  It helps audiences make informed data-driven decisions on key outcomes.

Here's what distribution visuals can show us:

1. **Shape of the distribution** - Distribution visuals allows us to observe the shape of the data distribution. Whether it is symmetric, skewed(right or left), bimodal(having two peaks) or multimodal(having multiple peaks), these shapes provide important information about the data's underlying pattern.

2. **Central tendency** - Distribution visuals hlep us identify the central tendency of data i.e. measures such as mean, median, and mode. Central tendency gives us a sense of the typical value in the dataset.

3. **Spread and Variability** - The spread of the distribution, as seen in measures like the range, interquartile range(IQR) and standard deviation, informs us about the variability or dispersion of the data ponts.

4. **Outliers and Anomalies** - Distribution visuals can reveal outliers. Outliers are data points that deviate significantly from the rest of the dataset. Identifying outliers is important for understanding anomalies or potential errors in the data.

## Histogram

- A histogram is a chart that shows the frequency or how often different things happen. It is a bar graph, but instead of showing individual values, it groups them into ranges.
- Each bar in a histogram represents a range of values, and the height of the bar shows how many times something falls within that range. 
- It helps us understand how data is spread out and what values are more common or less common. 
- Helps one see patterns in data quickly and easily.



Histograms are created using the *distplot/histplot* function in Seaborn. You only need to pass one argument which is the continous variable for which you would like to construct a histogram. 


https://www.youtube.com/watch?v=haJyaQObNwU



```python
peng = sns.load_dataset("penguins")
peng.head()
```

```python
sns.histplot(data=peng, x='bill_depth_mm', hue='island', palette='Paired')
plt.title("Distribution of Length of Penguin Bills")
plt.ylabel("Frequency")
```



Number of bins parameter - You specify the number of bins within the bin parameter. 
- If unspecified, Seaborn tries to find a useful number of bins to use.
- Increasing the number of bins generally increases the level of detail in the plot, providing a more granular view of the distribution. This may result in a visually more variable (noisier) plot.



Be thoughtful when choosing the number of bins as it might change the way you view your data.
By default, Seaborn determines the number of bins for the histograms based on the dataset. However, we can customize the number of bins.

```python
# Customizing the number of bins
sns.histplot(x='bill_length_mm',  data=peng,  hue='species', palette='tab20', bins=10)
```

## Kernel Density Plot

- Kernel Density Estimate(KDE) is a technique used to visualize the distribution of a continous variable. 

- A KDE creates a smooth curve that represents the distribution of data points in a dataset. It is like drawing a wavy line that shows the likelihood of occurence of different values in a dataset.

- Imagine having a bunch of dots on a piece of paper, and each dot represents a number. These dots could be anything like ages or temperatures. Now, instead of looking at the individual dots, you draw a smooth, wav line that goes up and down based on where they are more concentrated. 

- The wavy line is called a kernel density estimate. It helps you see patterns in the data. Where the line is higher, it means there are more dots there. It is like getting a general idea of where the values are most common and where they are rare.

- KDE is useful because it gives one a clear picture of how the data is spread out without having to focus on each individual dot. 

```python
tips_df = sns.load_dataset('tips')
tips_df.sample(7)
```

```python
# KDE plot for the total_bill column

sns.kdeplot(data=tips_df, x='total_bill')
```

```python
# KDE plot for the total_bill column

sns.histplot(data=tips_df, x='total_bill')
```

### Box Plot

A box plot aka whisker plot shows the distribution of data, which allows us to compare and visualize the data across different categories.

It consists of the following components:

1. Box - represents the Interquartile range(IQR), which is the range between the first quartile(Q1) and the third quartile(Q3). The length of the box indicates the spread of the middle 50% of the data. The median is shown as a line inside the box.

2. Whiskers - extend from the box to the minimum and maximum values within a defined range.

3. Outliers - are the individual data points that fall outside the whiskers.

### Violin Plots

A violin plot is a hybrid of a box plot and a kernel density plot. It shows peaks in the data. 

Unlike a boxplot that only shows summary statistics, violin plots depict summary statistics and the density of each variable. 


Components of a violin plot:
1. White dot - represents the median
2. Thick grey bar - represents the IQR.
3. Thin grey lines - represents the rest of the distribution





- The width of the violin represents the density of the data points at specific values. 

- Wider sections of the violin plot represent a higher probability that members of the population will take on the given value; the skinnier sections represent a lower probability.


```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.violinplot(data=iris, x='sepal_width')
```

```python
sns.violinplot(data=iris, x='species', y='sepal_width')
```

### Swarm Plot

- A swarm plot represents the distribution of data in the form of data points. 

- The data is sorted first, then we have an entry in the plot for each data point. The key feature is that  none of the points overlap each other. This creates a swarm of points that provide a clear view of the distribution and density of data within each category.

- It is recommended to draw a swarm plot complemented by a box or voilin plot.


```python

mpg_df = sns.load_dataset("mpg")
mpg_df.head()
```

```python
sns.swarmplot(x='horsepower', data=mpg_df)  
```

```python
mpg_df['horsepower'].value_counts()
```

```python
mpg_df['horsepower'].value_counts()
```

```python
# Using swarmplot to show distribution of the horsepower against the car origin

sns.swarmplot(x='horsepower',y='origin',data=mpg_df,size=3)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# distribution of ages by class

sns.swarmplot(y='age',x='class',data=titanic,size=3)
```

```python
titanic['age'].value_counts()
```

```python
sns.swarmplot(x='horsepower',y='origin', data=mpg_df, size=4, hue="cylinders", palette='Set1')
```

```python
# Seperate the strips
sns.set(rc={'figure.figsize':(10, 6)})
sns.swarmplot(x='horsepower',y='origin',data=mpg_df, size=4, hue="cylinders",palette='Set1', dodge=True)
```

```python
# Overlay Swarmplot with a box plot

sns.swarmplot(x='horsepower',y='origin',data=mpg_df,size=4, hue="cylinders",palette='rainbow')

sns.boxplot(x='horsepower', y='origin', data=mpg_df)
```

```python
# OVerlay the swarmplot with a violin plot
sns.swarmplot(x='displacement',y='origin',data=mpg_df,size=4, color='black')
sns.violinplot(x='horsepower', y='origin', data=mpg_df)              
```
