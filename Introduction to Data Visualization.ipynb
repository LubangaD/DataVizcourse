#### iii) Using the figure module

To draw a figure using the figure module, use the substeps below:

1. Create a canvas (figure) explicitly
2. Create a plot object by using add_subplot from the figure object.
3. Plot the chart by calling plot

```python
x = [21, 26, 38, 42]
y = [15, 21, 27, 38]

fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot()
ax1.bar(x,y)
```

The add_subplot method is used to add an Axes object to a Figure object. An Axes object represents a single plot or graph within a figure. You can use add_subplot to create multiple subplots in different arrangements.

#### Ticks
- Ticks are the values used to show specific points on the coordinate axis. 
- Are used to improve readability of plots.
- They can either be a number or a string.
- When we plot a graph, the axes adjust and take the default ticks. Matplotlib default ticks are generally sufficient in common situations but are in no way optimal for every plot.



The following are some properties that we can manipulate by setting parameters:

color: Sets the tick color
labelrotation: rotates the tick label
width: sets the tick width in points
length: sets the tick length in points
direction: sets the direction of the ticks, either towards the axis or away from it. Default value is out.


```python
import matplotlib.pyplot as plt

x = [21, 26, 38, 42]
y = [15, 21, 27, 38]
plt.plot(x, y)
# Tick customization
plt.tick_params(
    colors='green',        # Set tick color to red
    labelrotation=45,    # Rotate x-axis tick labels by 45 degrees
    width=2,             # Set tick width
    length=10,           # Set tick length
    direction='in'       # Set tick direction towards the axis
)
plt.show()
```

#### Line Plots

- Line plots help us understand relationships between numerical variables in our dataset.
- Visualize relationships over time


```python
import matplotlib.pyplot as plt


x = [21, 26, 38, 42]
y = [15, 21, 27, 38]

plt.plot(x,y, marker='o',markersize=8, linewidth=5, label='Line Plot')
plt.title('X vs Y')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.1)#alpha sets the transparency of the lines
plt.tick_params(
    colors='red',        # Set tick color to red
    labelrotation=90,    # Rotate x-axis tick labels by 45 degrees
    width=2,             # Set tick width
    length=10,           # Set tick length
    direction='out'       # Set tick direction towards the axis
)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
temperatures = [10, 12, 15, 20, 25, 30]
humidity = [65, 60, 55, 50, 45, 42]

# Line plot for temperatures
plt.plot(months, temperatures, label='Temperature', marker='o')

# Line plot for humidity
plt.plot(months, humidity, label='Humidity', marker='o')

# Adding title and labels
plt.title('Temperature and Humidity Variation (Jan-Jun)')
plt.xlabel('Months')
plt.ylabel('Values')

# Adding a grid with transparency
plt.grid(alpha=0.5)

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

## Box plot:
A box plot (or box-and-whisker plot) s is the visual representation of the depicting groups of numerical data through their quartiles against continuous/categorical data.

A box plot consists of 5 things.

1. Minimum
2. First Quartile or 25%
3. Median (Second Quartile) or 50%
4. Third Quartile or 75%
5. Maximum

[More On Box Plot] https://www.simplypsychology.org/boxplots.html

```python
# Plotting the box plot
sns.boxplot(x='Age', y='Weight', data=data)
```

## Violin Plot:
A violin plot is similar to a boxplot. It shows several quantitative data across one or more categorical variables such that those distributions can be compared. 

[More on Violin Plot]https://www.geeksforgeeks.org/violin-plot-for-data-analysis/
