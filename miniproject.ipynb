Since I found out there are null values, I'll have to clean the data set
-  Cleaning The Data Set:

Status and Unnamed1 columns have null values so I have to drop them:

```python
sales_cleaned=sales.drop(columns=['Status','unnamed1'])
sales_cleaned
```

The column 'Amount' has a few null values so I'll replace them with the median of the column
- First I calculate the median of the 'Amount' column
- Then replace the null values with the median


```python
amount_median=sales_cleaned['Amount'].median()

sales_cleaned['Amount'].fillna(amount_median, inplace=True)
```

# Heat Map

- Selecting the columns with numerals to craete a heatmap for correlation calculation 
- Creating the heatmap

```python
numerals_data = sales_cleaned.select_dtypes(include=[np.number])
correlation_matrix = numerals_data.corr()

fig = px.imshow(correlation_matrix,
                text_auto=True,  
                labels=dict(x="Features", y="Features", color="Correlation"),
                x=correlation_matrix.columns,
                y=correlation_matrix.columns)
fig.update_layout(title="Correlation Matrix Heatmap",
                  xaxis_title="Features",
                  yaxis_title="Features")

fig.show()
```

From the heatmap we can conclude that:
- Age and User_ID have a weak positive correlation which means that as age increases the User_ID increases in a small amount.ETC

## Bar Plot:Age Group vs Orders

- Creating a Bar plot of age groups vs orders so as to know the purchasing behavior among age groups

```python

fig = px.bar(sales_cleaned, x='Age Group',y='Orders',title='Bar plot for purchasing power',color_discrete_sequence=['blue'])

fig.update_layout(
    plot_bgcolor='lightgray', 
    paper_bgcolor='lightgray')  

fig.show()
```

From the bar plot we can conclude:
- The highest purchasing power is between the age group of 26-35
- The lowest purchasing power is between the age group of 0-17


## Pie Chart 

Checking states with the highest sales volume during the Diwali season

```python
fig = px.pie(sales_cleaned,values='Amount',names='State',title='States with the highest volume sales')

fig.show()
```

From the pie chart we can conclude that:
- The state of Uttar Pradesh has the highest number of volume sales
- The state of Telangana has the least number of volume sales

## Histogram

Creating a histogram to know the top selling product categories

```python
fig = px.histogram(sales_cleaned, x='Product_Category', color='Orders' ,nbins=5, title='Top selling product categories')

fig.show()
```

From the histogram we can conclude:
- Clothing and Apparel category is the top selling category with 659 people making 4 orders
- Hand and power tools category is the least selling category with only 12 people making 4 orders

## Box Plot

Creating a box plot to show how occupation influences spending patterns

```python
fig = px.box(sales_cleaned, x='Occupation', y='Amount', title='Influence of Occupation in Spending patterns')

fig.show()
```

From the box plot we can find out that:
- The government employees tend to spend more money since the maximum amount which has been used by them is above 23k almost reaching 24k 
- The people who have an occupation under Healthcare have the least amount spent which is a minimum of 188


## Scatter Plot

Creating a scatter plot showing spending between genders and how marital status affects it

```python
fig = px.scatter(sales_cleaned, x='Gender', y='Amount',color='Marital_Status', title='Influence of gender on spending with marital status')

fig.update_layout(
    width=800,  
    height=800   
)

fig.show()
```

From the scatter plot we can conclude that:
- The gender which has spent the most money is surprisingly the female gender for which they are not married (beating males by a very small mergin)
- Females also record the least amount of money spent spending as low as 188 and of course they are not married
