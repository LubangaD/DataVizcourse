DIWALI SALES DATA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
```

```python
diwali=pd.read_csv("Diwali Sales Data.csv",encoding=("latin1"))
diwali
```

```python
diwali.dropna(inplace=True)
```

```python
diwali.columns = diwali.columns.str.strip()
```

```python
fig=px.histogram(diwali,x= 'Age Group',y='Amount',color='Orders',title=' Purchasing Behavior  Across Different Age Groups')
fig.show()
```

Most purchases were made by people between age group 26-35

```python
plt.figure(figsize=(15,10))
group=diwali.groupby('Gender')['Amount'].size().reset_index()
plt.bar(group['Gender'],group['Amount'],color='skyblue')
plt.title('Gender vs Amount')
```

Most purchses were made by Females 

```python
fig=px.bar(diwali,x='Marital_Status',y='Orders')
fig
```

```python
plt.figure(figsize=(20,10))
sns.catplot(data=diwali,x='Marital_Status',y='Amount',hue='Product_Category',kind='bar')
plt.xlabel='Marital_Status'
plt.ylabel='Amount'
plt.title('How Marital status affect purchasing decision')
```

Auto was the most purchased item

```python
fig=px.pie(diwali, 'State','Amount',title='States with the higest sales ')
fig.show()
```

State with the highest sales was Uttar Pradesh

```python
plt.figure(figsize=(20,10))
sns.countplot(data=diwali,x='Zone',hue='Product_Category')
plt.xlabel='Zone'
plt.title('Difference in product preferences over different regions')
```

```python
fig=px.pie(diwali,'Product_Category','Amount', labels={'Product_Category': 'Product Category'},title='Product performance in different regions')
fig.show()
```

Food was the most purchased item

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Product_Category',y='Amount',hue='Zone',data=diwali)
plt.title('Product_Category vs Amount')
plt.style.use('default')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=diwali,x='Zone',hue='Product_Category')
plt.xlabel='Zone'
plt.title('Product performance in different regions')
```

Furniture was the most bought product

```python
fig=px.histogram(diwali,x= 'Occupation',y='Amount',color='Orders',title='How occupation affects spending')
fig.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=diwali, x='Orders', y='Amount', color='blue')
plt.title('Scatter Plot of Orders vs. Amount Spent')
plt.show()
```

```python
numcols =diwali.select_dtypes(include=[np.number])
corr_matrix = numcols.corr().drop(['User_ID'] , axis=0)
corr_matrix = corr_matrix.drop(['User_ID'],axis=1)
```

```python
fig = px.imshow(corr_matrix ,text_auto=True, 
                x = corr_matrix.columns , 
                y = corr_matrix.columns)
fig.update_layout(
    title = 'Heatmap showing Correlation between Age,Marital Status,Orders and Amount',
    xaxis_title = 'Features',
    yaxis_title = 'Features'
)
fig.show()
```
