# SCATTER PLOT:AGE VS PRODUCT CATEGORY

```python
#SCATTER PLOT:AGE VS PRODUCT CATEGORY
fig = px.scatter(df, x='Age', y='Product_Category')
fig.show()
```

# SCATTER PLOT:GENDER VS PRODUCT_CATEGORY

```python
#SCATTER PLOT:GENDER VS PRODUCT_CATEGORY
fig = px.scatter(df, x='Gender', y='Product_Category')
fig.show()
```

# Histogram: Gender vs Amount

```python
fig = px.histogram(df, x="Gender", y="Amount", color="Gender", title="Gender vs Amount")
fig.show()
```

```python
df.State.value_counts()
```

```python
df.State.unique()
```

```python
df_sales=df.groupby('State')['Amount'].sum().reset_index()
df_sales
```

```python
#BAR GRAPH:SALES BY STATE
df_sales = df_sales.sort_values(by='Amount', ascending=False)
fig = px.bar(df_sales, x='State', y='Amount', title='Sales by State')
fig.show()
```

```python
#PIE CHART:SALES BY STATE
df_sales_state=df.groupby('State')['Amount'].sum().reset_index()
fig=px.pie(df_sales_state,values='Amount',names='State',title='Sales by State')
fig.show()
```

```python
#PIE CHART:SALES BY PRODUCT_CATEGORY
df_sales = df.groupby('Product_Category')['Amount'].sum().reset_index()
fig = px.pie(df_sales, values='Amount', names='Product_Category', title='Sales by Product Category')
fig.show()
```

```python
df.Occupation.unique()
```

```python
#PIE CHART:SALES BY OCCUPATION
df_occupation_sales = df.groupby('Occupation')['Amount'].sum().reset_index()
fig = px.pie(df_occupation_sales, values='Amount', names='Occupation', title='Sales by Occupation')
fig.show()
```

# AGE GROUP VS SALES AMOUNT

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Age Group',data=df)
plt.show()
```

# Sales by Product Category for Age Group 0-17'

```python
sales_by_product_category = df[df['Age Group'] == '0-17'].groupby('Product_Category')['Amount'].sum().reset_index()
sales_by_product_category = sales_by_product_category.sort_values(by='Amount', ascending=False)
fig = px.bar(sales_by_product_category, x='Product_Category', y='Amount',title='Sales by Product Category for Age Group 0-17')
fig.show()
```

# **HEAT** **MAP**: ZONE VS AGE GROUP


```python
plt.subplots(figsize=(8, 8))
df_2= pd.DataFrame({x_label: grp['Zone'].value_counts()for x_label, grp in df.groupby('Age Group')})
sns.heatmap(df_2, cmap='viridis')
plt.xlabel('Age Group')
plt.ylabel('Zone')
```

**Eastern zone** has the **lowest** amount of **purchases overall** while **central zone** and **age group 26-35** has the **highest record of purchase**

# Eastern Region: Age vs Product Category

# Recommendations:

*   **Enhance Parental Marketing on the 0-17 Age Group:** Market products that are suitable for younger customers through parents, emphasizing educational value, safety, and long-term benefits to encourage purchases.


*   **Introduce Budget-Friendly Products:** Adding more affordable options in the product lineup, particularly in categories favored by younger consumers. This could include smaller packages or limited-time offers that appeal to their budget constraints.
*   **Expand Popular Product Lines:** Focus on expanding or enhancing product lines that are most popular among higher-spending age groups, ensuring that you maximize sales from segments with greater purchasing power.
